6. Explain the concept of generators in Python and how they are defined.
-> Generators in Python: Concept and Defination
1. What is a Generator?
-> A generator in python is a special type of function that returns an iterator and produces values one at a time using the yeild keyword instead of return.

-> Generator do not store all values in memory; instead, they generate values lazily (on demand). This makes them efficient for handling large datasets or infinite sequences.

2. Defining a Generator Function: Agenerator function is defined like a normal function but uses yield instead of return.


* Example: Simple Generator



```python
def simple_generator():
    yield 1
    yield 2
    yield 3

generator = simple_generator()
print(next(generator))
print(next(generator))
print(next(generator))
```

3. Difference Between yield and return
->
* return terminates the function and sends back a value.
* yield pauses the function and returns a value without ending it.


* Function with return run once and do not retain state.
* Function with yield retain their state and resume from where they paused.


* return returns a single value or a collection at once.
* yield produces values one at a time (lazy evaluation).


* return is memory-intensive for large data.
* yield id memory-efficient, as values are generatoed on demand.


4. Using Generators in a Loop: Instead of calling next() manually, you can use a for loop to iterate through a generator.


* Example: Using a Generator in a Loop:



















```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

for num in count_up_to(5):
  print(num)
```

7. What are the advantages of using generators over regular functions?
->

* Memory Efficiency:

-> Regular functions that return lists or other collections store all values in memory, which can be problematic for large datasets.

-> Generators, on the other hand, produce values one at a time, meaning they do not store the entire sequence in memory.

* Faster Execution:

-> A regular function must compute and store all values before returning them, while a gnerator computes only when next() is called.


* Lazy Evaluation (on-demand Execution):

-> Generators do not compute values in advance. Instead, they generate and return values only when requested.

-> For example, if you only need the first few results form a large dataset, a generator prevents unnecessary processing.


* State Retention:

-> Regular functions do not remember their previous state; they execute from the beginning each time thry are called.

-> Generators remember their last execution state and resume from where they left off when next() is called again.


* No need for large data storage:

-> If a function needs to return a sequence of values, it typically creates and returns a list.

-> If the list is too large, it can cause memory overflow.
No Need for Large Data Storage.


* Simpler Code (Easier to read and maintain)

-> Using yield in a generator eliminates the need for complex loops and manual state tracking.


* Efficient Iteration:

-> Generators integrate seamlessly with Python's for loops, avoiding the need for creating temporary lists.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
->


* map():

-> Purpose: Applies a function to each element in an iterable.

-> returns: A new iterable (map object) with transformed elements.

-> Use case: Data transformation (e.g., squaring numbers, formatting text).


* filter():

-> Purpose: Filters elements based on a condition (function returns True or False).

-> returns: A new iterable (filter object) with only matching elements.

-> Use case: Extracting specific elements (e.g., filtering even numbers)

* reduce():

-> Purpose: Repeatedly applies a function to reduce an iterable to a single value.

-> returns: A single accumulated result.

-> Use case: Aggregation (e.g., summing all elements).


* Example:











```python
from functools import reduce

numbers = [1, 2, 3, 4]
total = reduce(lambda x, y: x + y, numbers)
print(total)
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n < 2:
      return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % 1 == 0:
          return False
    return True

prime_numbers = [num for num in range(1, 201) if is_prime(num)]
print(prime_numbers)
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.



```python
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            fib = self.a + self.b
            self.a, self.b = self.b, fib
            self.count += 1
            return fib

fib_iter = FibonacciIterator(10)
for num in fib_iter:
    print(num, end=" ")
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(exponent):
    for i in range(exponent + 1):
      yield 2 ** i

max_exponent = 10
print(f"powers of 2 up to 2^{max_exponent}:")
for power in powers_of_two(max_exponent):
    print(power)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
filename = "sample_txt"

with open(filename, "w", encoding="utf-8") as file:
    file.write("Hello, this is a text file!\n")
    file.write("you can add multiple lines.\n")

print(f"File '{filename}' has been created successfully!")

```

```python
def read_file_line_by_line(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.strip()

for line in read_file_line_by_line("sample.txt"):
    print(line)
```
